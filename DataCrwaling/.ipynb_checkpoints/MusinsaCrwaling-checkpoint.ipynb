```python
import re

def clean_name(name):
  new_name = re.sub('[-=+,#/\?:^.@*\"※~ㆍ!』‘|\(\)\[\]`\'…》\”\“\’·]', ' ', name)
  new_name = ' '.join(new_name.split())
  return new_name
```

```python
imagePath = './상품이미지/'
headers = {'User-Agent': 'Mozilla/5.0'}
values = {'name': 'nesffer',
          'query': 'python'}
url = 'https://www.musinsa.com/category/'
def crwaling(categoryNumber, upperCategory, lowerCategory):
    main = urlopen(url+categoryNumber) # 카테고리 들어감
    bs_main = BeautifulSoup(main, "html.parser") # 해당 카테고리 페이지 읽음
    item_name, item_price, image_path, image_count = [], [], [], [] # 상품명, 가격, 이미지명
    savePath = imagePath+upperCategory+'/'+lowerCategory+'/' # 이미지 저장할 위치
    item_list = bs_main.findAll('li', {'class':'li_box'})
    count = 0
    for item in item_list:
        name = item.find('a', {'class':'img-block'}).get('title')
        name = clean_name(name.strip())
        item_name.append(name)

        price = item.find('p',{'class':'price'}).get_text().split()
        if len(price) == 1:
            item_price.append(price[0])
        elif len(price) == 2:
            item_price.append(price[1])

        item_url = item.find('a', {'class':'img-block'}).get('href') # 해당 상품 상세보기 url
        item_req = request.Request(item_url, headers=headers) # request
        item_html = urlopen(item_req) # 상세보기 들어감
        item_bs = BeautifulSoup(item_html, "html.parser") # 페이지 읽음

        # description = item_bs.find('p', {'class':'box_opinion_product'})
        # item_description.append(description) # p 태그 형태로 저장

        image_ul = item_bs.find('ul', {'class':'product_thumb'})
        image_li = image_ul.findAll('li')
        itemImage = []
        cnt = 0
        for i in range(len(image_li)): # 해당 상품 이미지 3개 다운로드
            img_url = image_li[i].find('img').get('src')
            if 'gif' in img_url:
                continue
            if 'peg' in img_url[-4:]:
                img, ext = 'https:'+img_url[:-7]+'500', img_url[-5:]
            else:
                img, ext = 'https:'+img_url[:-6]+'500', img_url[-4:]
            img_name = savePath+name+str(i+1)+ext
            try:
                request.urlretrieve(img+ext, img_name) # 상품명(번호).jpg
            except:
                print(f"[ERROR] file url: {img+ext} file name : {img_name}")
            itemImage.append(img_name)
            cnt += 1
            if cnt == 3: # 최대 3개
                break
        image_path.append(itemImage)
        image_count.append(cnt)
        count += 1
        print(f"{upperCategory}/{lowerCategory} : {count}/90")

    data = {'상품명':item_name, '가격':item_price, '이미지 경로':image_path, '이미지 개수': image_count}
    df = pd.DataFrame(data)
    return df
```

```python
df = crwaling('001001', '상의', '반팔')
df.to_csv("./상품DataFrame/test.csv")
```

```python
# 상의
top = {'반팔':'001001', '긴팔':'001010', '셔츠':'001002'}

# 바지
bottom = {'반바지':'003009', '슬랙스':'003008', '데님팬츠':'003002'}

# 스커트
skirt = {'미니스커트': '022001', '롱스커트': '022003'}

# 아우터
outer = {'롱패딩':'002013', '숏패딩':'002012', '코트': '002007', '트렌치 코트': '002008'}
```

```python
from collections import defaultdict

categoryDict = defaultdict(dict)
categoryDict['상의'] = top
categoryDict['바지'] = bottom
categoryDict['스커트'] = skirt
categoryDict['아우터'] = outer

for upper in categoryDict:
    for lower in categoryDict[upper]:
        df_name = './상품DataFrame/' + upper + '_' + lower + '.csv'
        # (categoryNumber, upperCategory, lowerCategory)
        df = crwaling(categoryDict[upper][lower], upper, lower)
        # print(upper, lower, categoryDict[upper][lower])
        df.to_csv(df_name)
```
